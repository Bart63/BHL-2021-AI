```python
subject = df['subject']
```

```python
df.drop(labels=['subject'], axis = 1, inplace = True)
```

```python
df.insert(0, 'Subject', subject)
```

## Cleaning

```python
mean_cols = [col for col in df.columns if 'mean' in col or 'Mean' in col]
std_cols = [col for col in df.columns if 'std' in col or 'Std' in col]

df_mean = df[mean_cols]
df_std = df[std_cols]

df_new = pd.concat([df_std, df_mean], axis = 1)
```

```python
angle_cols = [col for col in df_new.columns if 'angle' in col or 'Angle' in col]
energy_cols = [col for col in df_new.columns if 'energy' in col or 'Energy' in col]
meanFreq_cols = [col for col in df_new.columns if 'meanFreq' in col]
drop_cols = angle_cols + energy_cols + meanFreq_cols
drop_cols

df_new = df_new.drop(drop_cols, axis=1)
df_new = pd.concat([df.Activity, df_new], axis = 1)
df_new.insert(0, 'Subject', subject)
```

## Grouping

```python
grouped = df_new.groupby(['Subject', 'Activity'])
```

```python
df_grouped = grouped.mean()
df_grouped = df_grouped.reset_index()
df_grouped
```

## Save csv:

```python
df_grouped.to_csv('data/df_grouped.csv', index = False)
```
