## Wybieramy tylko te kolumny które uwzględniają średnią i odchylenie standardowe

```python
mean_cols = [col for col in df.columns if 'mean' in col or 'Mean' in col]
std_cols = [col for col in df.columns if 'std' in col or 'Std' in col]
```

```python
df_mean = df[mean_cols]
```

```python
df_std = df[std_cols]
```

```python
df_new = pd.concat([df_std, df_mean], axis = 1)
```

Dodajemy labele

```python
df_new = pd.concat([df.Activity, df_new], axis = 1)
```
